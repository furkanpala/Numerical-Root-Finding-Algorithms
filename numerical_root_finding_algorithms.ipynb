```python
# Bisection Method
import matplotlib.pyplot as plt
import numpy as np


def bisection(f,a,b,atol):
    if(a>b):
        print("Invalid interval")
        return
    
    if(f(a)*f(b) >= 0):
        print("Wrong guess")
        return
    
    n = int(np.ceil(np.log2(np.divide(b-a,2*atol))))
    
    p = a
    for i in range(n):
        p_new = (a+b)/2
        
        ax = plt.gca()
        plt.plot([p_new,p_new],[0, 600],"--",linewidth=1)
        
        error = np.abs((p_new - p) / p_new) * 100
        
        p = p_new
        
        print("iteration: %d, a: %f, b: %f, p: %f, f(a): %f, f(b): %f, f(p): %f, error: %%%.1f" %(i+1,a,b,p,f(a),f(b),f(p),error))
        

        if(f(p) == 0):
            break
            
        elif(f(a)*f(p) <0):
            b = p
            
        else:
            a = p
            
    
    return p


f = np.poly1d([1,-5,4])

a = -20
b = 1.1

bisection(f,a,b,0.001)

# Plot
x = np.arange(-600,10,0.01)
x_axis = x*0

plt.plot(x,x_axis)
plt.plot(x,f(x))
plt.plot(a,f(a),"go")
plt.plot(b,f(b),"go")
plt.axis([-21,2,-50,600])
```

```python
# Fixed Point Iteration Method
import matplotlib.pyplot as plt
import numpy as np

initial_guess = 3

x_iter = []
y_iter = []

f = np.poly1d([1,-5,4])
g = lambda x: (x**2 + 4) / 5

x_iter.extend([initial_guess,initial_guess])
y_iter.extend([0,g(initial_guess)])

def fixedPointIteration(f,g,x0,atol,maxiter):
    step = 0
    error = 100
    print("iteration: %d, x: %.2f, f(x): %.2f, error: - " %(0,x0,f(x0)))
    while(error > atol and step < maxiter):
        x1 = g(x0)
        
        error = np.abs((x1-x0)/x1)
        
        print("iteration: %d, x: %.2f, f(x): %.2f, error: %%%.2f" %(step+1,x1,f(x1),error*100))
        
        # To plot dashed iteration lines
        x_iter.extend([x0,x0,x1])
        y_iter.extend([x1,x1,x1])
        
        x0 = x1
        step = step+1
    
    return x0




fixedPointIteration(f,g,initial_guess,0.001,200)


# Plot
x = np.arange(-10,10,0.01)
plt.plot(x,g(x),label="g(x)")
plt.plot(x,x,"r",label="y = x")
plt.plot(x_iter,y_iter,"g--")
plt.axis([0,5,0,5])
plt.legend()
```

```python
# Newton's Method
import numpy as np
import matplotlib.pyplot as plt


x = np.arange(-60,60,0.01)
f = np.poly1d([1,-8,12])
df = np.poly1d([2,-8])



def line(x1,y1):
    return ((df(x1)*x1) -y1)/df(x1)


def NewtonMethod(f,df,x0,atol):
    step = 0
    
    error = 100
    
    print("iteration: %d, x: %f, f(x): %f, df(x): %f, error: -" %(0,x0,f(x0),df(x0)))
    
    
    plt.scatter(x0,f(x0),color="C1",s=50)
    plt.plot([x0,line(x0,f(x0))],[f(x0),0],"--",linewidth=1.2,label="df n=%d" %(step))
    plt.annotate(f"$x_{step}$",xy=(x0,0),xytext=(x0,-1))
    plt.plot([x0,x0],[0,f(x0)],"b--")

    while(error>atol):
        x1 = x0 - (f(x0) / df(x0))
        
        error = np.abs((x1-x0)/x1)
        
        print("iteration: %d, x: %f, f(x): %f, df(x): %f, error: %%%f" %(step+1,x1,f(x1),df(x1),error*100))
        
        if(step < 2):
            plt.scatter(x1,f(x1),color="C1",s=50)
            plt.plot([x1,line(x1,f(x1))],[f(x1),0],"--",linewidth=1.2,label="df n=%d" %(step+1))
            plt.annotate(f"$x_{step+1}$",xy=(x1,0),xytext=(x1,-1))
            plt.plot([x1,x1],[0,f(x1)],"b--")
        
        x0 = x1
        
        step = step+1
    return x0,step



NewtonMethod(f,df,4.5,0.001)

plt.plot(x,f(x),"black",label="f(x)",linewidth=3)
plt.plot(x,x*0)
plt.axis([4,8.5,-7,15])
plt.legend()
```

```python
# Secant Method
import numpy as np
import matplotlib.pyplot as plt

def secant(f,x0,x1,atol,maxiter):
    step = 0
    x2 = x1
    error = 100
    
    plt.plot([x0,x1],[f(x0),f(x1)],"b",linewidth=0.75,label="n=0")
    plt.plot([x0,x0],[0,f(x0)],"--")
    plt.annotate(f"$x_{0}$",xy=(x0,0),xytext=(x0,-1))
    plt.plot([x1,x1],[0,f(x1)],"--")
    plt.annotate(f"$x_{1}$",xy=(x1,0),xytext=(x1,-1))
    
    
    while(error>atol and step < maxiter):
        x2 = x1 - ((f(x1)*(x1-x0))/(f(x1)-f(x0)))
        
        error = np.abs((x2 - x1) / x2)
        
        print("iteration: %d, x: %f, f(x): %f, error: %%%f" %(step+1,x2,f(x2),error*100))
        
        x0 = x1
        x1 = x2
        
        if(step < 1):
            plt.plot([x0,x1],[f(x0),f(x1)],"b",linewidth=0.75,label="n=1")
            plt.plot([x1,x1],[0,f(x1)],"--")
            plt.annotate(f"$x_{2}$",xy=(x1,0),xytext=(x1,-1))
        
        step = step+1
    
    return x1

f = np.poly1d([1,-5,4])
secant(f,-5,5,1e-6,100)

x = np.arange(-50,50,0.001)
plt.plot(x,f(x))
plt.plot(x,x*0)
plt.axis([-6,8,-5,60])
plt.legend(loc="best")
```

```python
# Third question
import numpy as np

dataset = np.genfromtxt("cit-HepTh.txt",delimiter="\t",skip_header=4,dtype=int)

nodes = {}

adjMatrix = np.zeros((27770,27770),dtype=int)

i = 0

for data in dataset:
    lNode = data[0]
    rNode = data[1]
        
    if(nodes.get(lNode) == None):
        nodes[lNode] = i
        i = i + 1
    
    if(nodes.get(rNode) == None):
        nodes[rNode] = i
        i = i + 1
        
    lNodeID = nodes[lNode]
    rNodeID = nodes[rNode]
    
    adjMatrix[lNodeID][rNodeID] = 1
    
sumColumns = np.sum(adjMatrix,axis = 0)
topTen = sumColumns.argsort()[-10:][::-1]

# Inverse dictionary
nodes = {v:k for k,v in nodes.items()}

for i in topTen:
    print("ID:",nodes[i],"Citation amount:",sumColumns[i])
```
